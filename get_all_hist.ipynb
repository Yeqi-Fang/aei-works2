```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
jsons = os.listdir("LAL_example_data/LALSemiCoherentF0F1F2_aggressive_memory")
jsons = [j for j in jsons if j.endswith(".json")]
jsons.sort()
# jsons
```

```python
json_test = 'LAL_example_data/LALSemiCoherentF0F1F2_aggressive_memory/config_mismatch_run_0001.json'
import json
with open(json_test, 'r') as f:
    data = json.load(f)
mf = data['config']['mf']
mf1 = data['config']['mf1']
mf2 = data['config']['mf2']
T_coh = data['config']['T_coh']
gamma1 = data['config']['gamma1']
gamma2 = data['config']['gamma2']
mismatches = data['mismatch_list']
mean_mismatch = data['statistics']['mean_mismatch']
```

```python
def plot_hist(mismatches, mf, mf1, mf2, T_coh, gamma1, gamma2, N, depth):
    fig, ax = plt.subplots(figsize=(10, 6))
    mean_mismatch = np.mean(mismatches)
    ax.hist(
        mismatches,
        bins=20,
        density=True,
        color="#5B9BD5",
        alpha=0.85,
        edgecolor="black",
        linewidth=1.0,
    )
    
    ax.set_xlabel(r"mismatch $\mu$", fontsize=20)
    ax.set_ylabel("normalized histogram", fontsize=20)
    ax.set_xlim(0, 1)
    ax.tick_params(axis="both", which="major", labelsize=14, length=6)
    ax.grid(axis="y", linewidth=0.6, alpha=0.35)
    ax.set_title(f"Mismatch Distribution (N={len(mismatches)}, μ={mean_mismatch:.4f})", fontsize=16)
    
    fig.tight_layout()
    
    # 保存图片
    os.makedirs("images", exist_ok=True)
    fig.savefig(f"images/MC2/mismatch_distribution_aggressive_{mf}-{mf1}-{mf2}-{T_coh}-{gamma1}-{gamma2}-{N}-{depth}.pdf")
    fig.savefig(f"images/MC2/mismatch_distribution_aggressive_{mf}-{mf1}-{mf2}-{T_coh}-{gamma1}-{gamma2}-{N}-{depth}.png")
    print(f"Histogram saved to images/MC2/mismatch_distribution_aggressive_{mf}-{mf1}-{mf2}-{T_coh}-{gamma1}-{gamma2}-{N}-{depth}.pdf")
    
    plt.close(fig)
    # plt.show()
```

```python
for json_file in jsons:
    with open(f"LAL_example_data/LALSemiCoherentF0F1F2_aggressive_memory/{json_file}", 'r') as f:
        data = json.load(f)
    
    mf = data['config']['mf']
    mf1 = data['config']['mf1']
    mf2 = data['config']['mf2']
    T_coh = data['config']['T_coh']
    gamma1 = data['config']['gamma1']
    gamma2 = data['config']['gamma2']
    mismatches = data['mismatch_list']
    
    N = len(mismatches)
    depth = 0.6  # Assuming depth is not provided in the JSON, set to 0 or adjust as needed
    
    plot_hist(mismatches, mf, mf1, mf2, T_coh, gamma1, gamma2, N, depth)
```
