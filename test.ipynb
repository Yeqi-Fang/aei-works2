```python
import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(-3.14, 3.14, 1000)
b = (np.sin(4 * a) / (4 * a + 1e-8)) ** 2

plt.plot(a, b, label='$(\sin(x)/x)^2$')
plt.title('Plot of $(\sin(x)/x)^2$')
plt.xlabel('x')
plt.ylabel('$(\sin(x)/x)^2$')
plt.ylim(0, 1.1)
plt.show()
```

```python
c = np.random.uniform(-3.14, 3.14, 10000)

d = 1 - (np.sin(c) / (c )) ** 2

plt.hist(d, bins=30, density=True, alpha=0.7, label='Histogram of $(\sin(x)/x)^2$')
```
